# Meat consumption until 2019 
apro_mt_pheadm_1_Data.csv 

```python
meat_consumption_path = path + 'apro_mt_pheadm_1_Data.csv'
meat_consumption = pd.read_csv(meat_consumption_path)
meat_consumption = meat_consumption.drop(columns = ['Flag and Footnotes'])

meat_consumption['Value'] = meat_consumption['Value'].str.replace(':', '0', regex=True)

meat_consumption = meat_consumption[meat_consumption['UNIT'] == 'Thousand tonnes']
meat_consumption = meat_consumption.drop(columns= ['UNIT'])

# Slaughtering only
meat_consumption = meat_consumption[meat_consumption['MEATITEM'] == 'Slaughterings']
meat_consumption = meat_consumption.drop(columns= ['MEATITEM'])
```
